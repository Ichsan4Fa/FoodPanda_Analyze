#Load Dataset

Dataset loaded from kaggle connected by kagglehub library

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("ayeshaimran1619/foodpanda-data-analysis")

print("Path to dataset files:", path)
```

Finding the correct name files

```python
import os
files = os.listdir(path)
for f in files:
  print(files)
```

Load dataset using pandas read_csv()

```python
import pandas as pd
import numpy as np

df = pd.read_csv(path + "/Foodpanda Analysis Dataset.csv")
df.head()
```

```python
kfc_cities = df[df['restaurant_name'] == 'KFC']['city'].unique()
num_kfc_each_cities = df[df['restaurant_name'] == 'KFC'].groupby('city').size()
print("Cities with KFC restaurants:", kfc_cities)
print("Cities with KFC restaurants:", num_kfc_each_cities)
```

```python
import matplotlib.pyplot as plt

delivery_status_counts = df['delivery_status'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(delivery_status_counts, labels=delivery_status_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Delivery Status')
plt.axis('equal')
plt.show()
```
